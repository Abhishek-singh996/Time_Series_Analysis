## Time resampling

```python
df_tesla.resample(rule='A').max()
```

```python
## year end freq
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
## quarterly
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
## Business End Frequency
df_tesla.resample(rule='BA').max()
```

```python
df_tesla['Open'].resample(rule='BA').max().plot(kind='bar')
```

```python
df_tesla['High'].rolling(10).mean().head(20)
```

```python
df_tesla['High: 30 days rolling'] = df_tesla['High'].rolling(30).mean()
```

```python
df_tesla.head()
```

```python
df_tesla[['High','High: 30 days rolling']].plot(figsize=(12,6))
```

## Microsoft Data using pandas data reader

```python
import yfinance as yf
import pandas_datareader.data as pdr
yf.pdr_override()
df_msft = pdr.get_data_yahoo('MSFT')
```

```python
df_msft.tail()
```

```python
df_msft.info()
```

```python
df_msft.High.plot(xlim=['2017-01-01','2022-12-31'],ylim=[0,600],figsize=(12,4),ls='--',c='green')
```

```python

df_msft['High'].resample(rule='A').max().max()
```

```python

df_msft.loc['2017-01-01' : '2023-01-01'].High.resample('A').max().plot()
print('max of stock price is: ',df_msft.loc['2017-01-01' : '2023-01-01'].High.resample('A').max().max())
```

```python
print(df_msft[df_msft.High == df_msft.High.max()].index)
```

```python
print(df_msft[df_msft.Low == df_msft.Low.min()])
```
